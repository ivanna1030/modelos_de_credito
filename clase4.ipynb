# Modelo de puntuación de crédito FICO.

**Introducción al modelo de puntuación de crédito FICO:**

- El modelo de puntuación de crédito FICO es un sistema desarrollado por Fair Isaac Corporation que se utiliza para evaluar el riesgo crediticio de las personas. FICO genera una puntuación de crédito de entre 300 y 850 puntos basada en la información de los informes de crédito de una persona. Cuanto más alta es la puntuación FICO, menor es el Riesgo percibido y viceversa.

**Los 5 factores principales:**

- **El historial de pago:** Si el prestatario ha pagado sus facturas y préstamos anteriores a tiempo.

- **Cantidad adeudada:** La cantidad total de deuda que tiene el prestatario.

- **Historial de crédito:** Cuánto tiempo ha tenido el prestatario líneas de crédito abiertas.

- **Nuevas solicitudes de crédito:** Cuántas nuevas solicitudes de crédito ha hecho el prestatario recientemente.

- **Tipos de crédito:** La mezcla de tipos de crédito (tarjetas de crédito, préstamos, etc.) que tiene el prestatario.
 
Consejos para mejorar tu puntuación:

- **Pagar facturas a tiempo:** Pagar todas las facturas antes de la fecha de vencimiento ayuda a mejorar el puntaje de crédito.

- **Mantener los saldos bajos en tarjetas de crédito:** Tener menos del 30% del límite de crédito utilizado en cada tarjeta de crédito mejorará el puntaje.

- **No solicitar crédito nuevo frecuentemente:** Cada solicitud de crédito nuevo reduce temporalmente el puntaje de crédito.

- **Revisar reportes de crédito:** Revisar los reportes de crédito periódicamente para asegurarse de que no haya errores.

- **Construir historial de crédito:** Tener varios tipos de crédito (tarjetas de crédito, préstamos, etc) por varios años ayuda a mejorar el puntaje.

**Críticas al modelo FICO... algunas:**

- **Falta de transparencia:** El algoritmo FICO es propietario y no se conoce públicamente.

- **Sesgo racial:** El modelo FICO puede reproducer sesgos raciales debido a disparidades socioeconómicas históricas.

- **No predice comportamiento futuro:** El modelo FICO se basa en el historial crediticio pasado y no necesariamente predice el comportamiento futuro del prestatario.

- **Ignora factores cualitativos:** El modelo FICO no considera factores cualitativos como estabilidad laboral, nivel educativo, etc.

- **Dificultad para mejorar el puntaje:** Puede ser difícil para algunos prestatarios mejorar significativamente su puntaje FICO en el corto plazo.
