# <span style="color: rgb(146, 66, 94);">Altman Z-score and Merton Model</span>

## <span style="color: rgb(191, 146, 162);">Altman Z-score</span>

<span style="background-color: rgb(219, 182, 195);">What is the Altman Z-score?</span>

Altman Z-score is a financial analysis tool developed in 1968 by Edward Altman to forecast the likelihood of a publicly traded company facing bankrupt within two years. It evaluates a company's financial stability by using five financial ratios: working capital, retained earnings, EBIT, market value of equity, and sales relative to total assets. This way, it helps investors and lenders decide whether a company is financially reliable and capable of repaying its debts before they choose to invest in it or lend money to it.

<span style="background-color: rgb(219, 182, 195);">Formula</span>

$$Z=1.2x_1+1.4x_2+3.3x_3+0.6x_4+1.0x_5$$

**where:**

- $x_1 = \frac{\text{Working Capital}}{\text{Total Assets}}:$ It measures a company's liquidity.

- $x_2 = \frac{\text{Retained Earnings}}{\text{Total Assets}}:$ It measures profitability.

- $x_3 = \frac{\text{EBIT}}{\text{Total Assets}}:$ It measures operating efficiency.

- $x_4 = \frac{\text{Market Cap}}{\text{Total Liabilities}}:$ It measures leverage.

- $x_5 = \frac{\text{Revenue}}{\text{Total Assets}}:$ It measures asset turnover.

<span style="background-color: rgb(219, 182, 195);">Interpretation</span>

- $Z \lt 1.8:$ The company is in the financial distress zone, with a high possibility of bankruptcy.

- $1.8 \leq Z \lt 3:$ The company is in the financial grey zone, with a moderate possibility of bankruptcy.

- $Z \geq 3:$ The company is in the financial safe zone, with a low possibility of bankruptcy.

<div style="height: 60px;"></div>

<span style="background-color: rgb(219, 182, 195);">Applications</span>

- **Financial condition:** The Z-Score reflects a company’s financial health and estimates the likelihood of bankruptcy.

- **Investment choices:** Investors rely on the Z-Score to measure potential risk before deciding whether to invest.

- **Credit decisions:** Lenders use the Z-Score to evaluate a company’s ability to repay debt and determine whether to grant financing.

<span style="background-color: rgb(219, 182, 195);">Limitations</span>

- **Ignores qualitative factors:** The Altman Z-Score relies only on numerical financial ratios and does not consider key qualitative elements such as management quality, industry dynamics, or competitive strength.

- **Industry limitations:** Since the original model was created using data from manufacturing companies, its accuracy may decrease when applied to companies in other industries.

- **Based on past data:** The model uses historical financial information, which means it may fail to capture unexpected future changes or sudden financial difficulties.

- **Limited use for private companies:** The model was developed for publicly traded companies with accessible financial reports, so its application to private companies can be restricted due to limited disclosure.

<span style="background-color: rgb(219, 182, 195);">Versions</span>

Altman created three different versions of the Z-score to suit different types of companies. The original model, introduced in 1968, was designed for publicly traded manufacturing companies with total assets greater than $1 million. This original formula did not apply to private companies or to non-manufacturing companies with assets below that threshold. Later, in 1983, he introduced two additional variations for smaller private manufacturing companies. Model A was created specifically for private manufacturing companies, while Model B was created for non-publicly traded companies.

However, the 1968 model was used in this analysis because it is the original and most widely recognized version of the Altman Z-score. Although the S&P 500 includes both manufacturing and non-manufacturing companies, many of which have total assets above $1 million, the original formulation remains the benchmark model in financial analysis and provides a standardized framework for assessing bankruptcy risk.

## <span style="color: rgb(191, 146, 162);">Merton Model</span>

<span style="background-color: rgb(219, 182, 195);">What is the Merton Model?</span>

The Merton Model is an important tool introduced in 1974 by Robert Merton for measuring a company's probability of default. It treats a company’s equity as a call option on its underlying assets, allowing analysts and lending professionals to evaluate credit risk more effectively. By understanding this model, financial practitioners can better assess structural credit risk and make more informed financial decisions.

The model uses three main components: the total market value of the company’s assets (V), the face value of its debt that matures at a specific future date (D), and equity, which is treated as a call option on the company’s assets with a strike price equal to the face value of the debt. At maturity, if the value of the assets exceeds the debt (V > D), the company repays its obligations and shareholders keep the residual value (V − D). However, if the asset value falls below the debt level (V < D), the company defaults, shareholders receive nothing, and control of the company passes to creditors.

Liabilities represent the company's debt with a specific maturity T and act as the strike price in the option framework. Default occurs if the asset value falls below L at maturity. In practice, companies have multiple debt maturities, so a liability threshold (or default point) is used, typically set between short-term liabilities and total liabilities. In this analysis, for a more conservative estimation of the probability of default, total debt is used as the liability threshold.

<span style="background-color: rgb(219, 182, 195);">Formula</span>

$$E = V_t N(d_1) - K e^{-r(T - t)} N(d_2)$$

**where:**

- $d_1 = \frac{\ln\left(\frac{V_t}{K}\right) + \left(r + \frac{1}{2}\sigma^2\right)(T - t)}{\sigma \sqrt{T - t}}$

- $d_2 = d_1 - \sigma \sqrt{T - t}$

- $E$ = Theoretical value of the company’s equity

- $V_t$ = Value of the company’s assets at time $t$

- $K$ = Face value of the company’s debt

- $t$ = Current time

- $T$ = Future maturity time

- $r$ = Risk-free interest rate

- $N(\cdot)$ = Cumulative standard normal distribution

- $e$ = Exponential constant

- $\sigma$ = Volatility (standard deviation) of asset returns

<span style="background-color: rgb(219, 182, 195);">Credit risk measurement</span>

- **Distance to Default (DD):** Number of standard deviations between expected asset value at debt maturity and the liability threshold. A lower distance to default indicates a higher probability of default.

    - $DD = \frac{\ln(\frac{V}{D}) + (r + \frac{\sigma^2}{2})T}{\sigma\sqrt{T}}$

- **Probability of Default (PD):** Probability of the asset value falling below the liability threshold.

    - $PD = 1 - N(DD)$

<span style="background-color: rgb(219, 182, 195);">Interpretation</span>

- $PD \lt 5\%:$ The company has a low risk of default.

- $5\% \leq PD \lt 20\%:$ The company has a moderate risk of default.

- $PD \geq 20\%:$ The company has a high risk of default.

<span style="background-color: rgb(219, 182, 195);">Assumptions</span>

- **Continuous-time setting:** The company’s asset and debt values change continuously.

- **Geometric Brownian motion:** The company’s asset value follows a stochastic process consistent with geometric Brownian movement.

- **Tradable securities:** Both equity and debt are actively traded in financial markets.

- **Risk-neutral valuation:** Market participants are treated as risk-neutral, and valuation is performed by discounting expected payoffs at the risk-free rate.

- **No market frictions:** Taxes, transaction costs, and other market imperfections are not considered.

- **Constant volatility:** The volatility of the company’s asset value remains stable over time.

<span style="background-color: rgb(219, 182, 195);">Applications</span>

- **Credit risk assessment:** The model estimates the probability that a company will default by comparing asset value to debt obligations at maturity.

- **Distance to default measurement:** It quantifies how far a company’s asset value is from the default threshold in standard deviation terms.

- **Bankruptcy and financial distress analysis:** The framework evaluates corporate solvency by analyzing asset levels, debt structure, and maturity.

- **Credit spread estimation:** The model allows investors to derive implied credit spreads that reflect default risk.

- **Asset volatility estimation:** It enables practitioners to infer unobservable asset value and volatility from observable equity data.

<span style="background-color: rgb(219, 182, 195);">Limitations</span>

- **Limited applicability:** The model is not well suited for unsecured loans (such as microcredit or credit cards) or for non-traded markets like mortgages.

- **Default definition restriction:** It treats insolvency as the sole cause of default, ignoring other factors such as liquidity problems or debt service difficulties.

- **Weak empirical performance:** The model’s forecasts may not align closely with observed real-world default data.

- **Incomplete risk factor integration:** There is no straightforward way to incorporate certain risk drivers, such as changes in interest rates.

## <span style="color: rgb(191, 146, 162);">Credit approval</span>

In this analysis, a credit approval function was created based on both the Altman Z-score and the Merton Model probability of default.

The decision rule follows a conservative approach: a company is approved for credit only if it demonstrates financial strength under both models simultaneously. Specifically, companies with a Z-score above 1.8 are considered to be outside the financial distress zone, while companies with a probability of default below 20% are considered to have an acceptable level of market-implied insolvency risk. If both conditions are satisfied, the company is approved for credit; otherwise, the application is denied.

This dual-criteria approach reduces model risk by combining accounting-based measures of solvency with market-based measures of asset volatility and default risk. By requiring consistency across both frameworks, the decision rule ensures a more robust and risk-averse credit evaluation process.

## <span style="color: rgb(191, 146, 162);">Credit risk analysis</span>

For this analysis, both models were applied to three stock market companies; Apple Inc, The Walt Disney Company and Warner Bros Discovery Inc.

<span style="background-color: rgb(219, 182, 195);">About the companies</span>

<img src="market.png" width="650">

As we can see, the three companies differ significantly in size and market risk exposure.

Apple is the largest company, with a market capitalization of $3.89 trillion, reflecting strong financial stability and global dominance in the technology sector, while its beta of 1.11 indicates moderate volatility slightly above the market average.

Disney is considerably smaller at $187.23 billion and exhibits a higher beta of 1.42, suggesting greater sensitivity to overall market fluctuations and higher cyclical exposure.

Warner Bros is the smallest company at $71.29 billion and presents the highest beta of 1.62, indicating substantial volatility and increased market risk.

Overall, based on market indicators alone, Apple appears to be the most stable company, Disney occupies an intermediate position, and Warner Bros exhibits the highest volatility and valuation uncertainty. These characteristics are particularly relevant when applying the Merton model, as higher equity volatility may translate into higher asset volatility and a greater probability of default.

<span style="background-color: rgb(219, 182, 195);">Altman Z-score</span>

<img src="altman2.png" width="650">

As we can see, for AAPL, the slightly negative X1 (-0.05) indicates that working capital is relatively low compared to total assets, which may reflect efficient balance sheet management rather than distress. The small negative X2 (-0.04) suggests limited reliance on accumulated retained earnings relative to asset size, possibly due to dividend payouts or share buybacks. However, X3 is strong (0.37), showing solid operating profitability, and X4 is exceptionally high (13.62), meaning the market value of equity vastly exceeds total liabilities, signaling very low leverage risk and strong investor confidence. X5 (1.16) reflects efficient use of assets to generate sales. Together, these factors explain the very high Z-score (10.44), placing the company in the safe zone.

For DIS, the negative X1 (-0.05) points to tighter liquidity conditions. In contrast, X2 (0.31) shows a meaningful level of retained earnings relative to assets, indicating some historical profitability. X3 (0.07) reveals moderate operating performance, weaker than Apple’s but still positive. X4 (2.26) suggests that equity exceeds liabilities, though with less cushion than AAPL, implying moderate leverage risk. X5 (0.48) indicates lower asset turnover efficiency. These mixed but generally stable indicators justify its overall Z-score of 2.43, placing the company in the grey zone.

For WBD, the negative X1 (-0.02) indicates limited short-term liquidity, which becomes more concerning when combined with weaknesses in other financial dimensions. The negative X2 (-0.12) suggests accumulated losses or limited retained earnings, reflecting weaker long-term financial stability. The negative X3 (-0.09) is particularly critical, as it indicates operating losses relative to total assets, signaling poor profitability performance. X4 (1.02) shows that the market value of equity only slightly exceeds total liabilities, implying elevated leverage risk and a minimal financial cushion against potential distress. X5 (0.38) reflects relatively low asset efficiency in generating sales. The combination of weak profitability, constrained liquidity, and limited equity buffer explains the low Z-score (0.51), placing the company in the distress zone.

<span style="background-color: rgb(219, 182, 195);">Merton Model</span>

<img src="merton.png" width="650">

As we can see, for AAPL, the equity value and asset value are identical ($3,982.62B), which suggests the model assumes no significant debt or that liabilities are already embedded in the valuation framework. Asset volatility is moderate (31.37%), reflecting normal market fluctuations for a large technology company. The default probability is effectively 0.00%, indicating an extremely strong financial position and very low credit risk. The combination of massive company value and negligible default probability supports its classification as low risk.

For DIS, equity and asset values are also equal ($229.92B), implying a similar structural modeling assumption. Asset volatility is lower than AAPL’s (26.04%), suggesting relatively more stable asset dynamics. The default probability is again 0.00%, meaning the model does not detect meaningful short-term default risk. Despite Disney’s earlier grey zone Z-score, this structural model signals financial stability, justifying its classification as low risk.

For WBD, although the equity and asset values match ($108.67B), asset volatility is the highest among the three companies (38.47%), indicating greater uncertainty in company value. Its default probability (0.51%) is small in absolute terms but noticeably higher than AAPL and DIS. This suggests that while the company is still categorized as low risk, it carries relatively higher credit risk compared to the other two companies. The higher volatility combined with a non-zero default probability reflects greater financial fragility.

<span style="background-color: rgb(219, 182, 195);">Decision</span>

<img src="summary.png" width="650">

<img src="both2.png" width="650">

As we can see, when combining both the Altman Z-score and the Merton model results, AAPL clearly demonstrates financial strength. Its very high Z-score (10.44) places it in the safe zone, reflecting strong liquidity, profitability, and low leverage, while the Merton model reports an asset value of $3982.62B, moderate volatility (31.37%), and a 0.00% probability of default, confirming low risk. Together, these indicators justify the approved credit decision.

For DIS, the Z-score (2.43) falls in the grey zone, signaling moderate financial health with some leverage and lower operating efficiency compared to AAPL. However, the Merton model shows an asset value of $229.92B, volatility of 26.04%, and a 0.00% probability of default, also low risk. When combining both models, the company shows moderate caution but overall stability, supporting a conservative approved decision.

For WBD, the Z-score (0.51) is in the distress zone, indicating weak liquidity, low profitability, and higher leverage risk. The Merton model shows an asset value of $108.67B, higher volatility (38.47%), and a 0.51% probability of default, still technically low risk. When combining both models, the very low Z-score outweighs the low market-implied default, leading to a denied credit decision.
