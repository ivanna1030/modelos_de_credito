# Deconstructing Credit Rates: Analysis & Valuation.

The Foundation: The Building Block Model of Interest Rates.

**The Master Formula:**

- The nominal interest rate r is never a single number; it is a sum of compensation for distinct risks.

    - $r=R_{real}+\pi_{exp} +CRP+DRP+LP+MRP$

- **Real Rate ($R_{real}$):** Compensation for deferring consumption.

- **Inflation ($\pi_{exp}$):** Maintaining purchasing power Inflation.

- **Risk Premiums:** Extra return for specific dangers (Country, Default, Liquidity).

 

**Factor 1: Country Differences.**

**Why borrowing in Brazil costs more than in Japan:**

**Drivers of Cross-Border Variance:**

- **The Fisher Effect:** Nominal rates must cover expected inflation. If Country A expects 5% inflation and Country B expects 1%, Country A's base rate must be ~4% higher simply to maintain Parity.

    - $(1+i)=(1+R_{real})×(1+\pi_{exp})$

- **Sovereign Risk (CRP):** The spread investors demand to hold a nation's debt over "risk-free" US Treasuries. It reflects political stability, deficit control, and currency volatility.

    - **Example:** Brazil Bond Yield - US Treasury Yield = CRP


**Global Base Rate Comparison:** The massive gap between Japan and Brazil is not arbitrary; it is the mathematical sum of Inflation Expectations ($\pi$) and the Country Risk Premium (CRP).

**Factor 2: Product Differences.**

**Mortgages vs. Credit Cards:**

**Credit Risk Decomposition:**

- **Why is a mortgage 6% and a credit card 20%?** It comes down to Loss Given Default (LGD).

- **Secured Loans (Mortgage):** The house is collateral. LGD is low (~20%). If you default, the bank sells the home.

- **Unsecured Loans (Cards):** No collateral. LGD is high (~100%). If you default, the money is gone. Lenders charge a massive premium to offset this "Expected Loss".
 

**The Cost of Embedded Options: Lenders charge extra when the borrower holds an "Option" that disadvantages the bank.**

- **Prepayment Risk (Call Option):** In mortgages, borrowers can refinance when rates drop. The bank loses the high-interest asset exactly when it's most valuable. Rates include a premium for this risk.

- **Liquidity Premium (Lines of Credit):** Revolving credit gives borrowers the right to draw funds. Banks must keep cash idle and ready, earning nothing. A fee is charged for this "on-demand" liquidity.


**Factor 3: Bank Differences.**

**Funds Transfer Pricing & Regulation:**

**Why Banks Price Differently:**

- **Funds Transfer Pricing (FTP):** The bank's internal "Cost of Goods Sold". A bank rich in cheap deposits (0.1% cost) has a lower FTP than a bank relying on expensive bond markets (5% cost).

- **Regulatory Capital (Basel III):** Banks must hold expensive Equity against risky loans. A bank with a weak balance sheet pays a higher "Capital Charge," forcing them to hike loan rates to maintain ROE.
