# Merton Model KMV

**Fundamentos del Modelo de Merton:**

- El modelo de Merton se basa en la teoría de opciones y considera la estructura de capital de una empresa para determinar la probabilidad de incumplimiento. En esencia, el modelo ve el capital de una empresa como una opción de compra sobre sus activos.

**Estructura básica:**

- **Activos:** El valor total de los activos de la empresa.

- **Pasivos:** El valor facial de la deuda de la empresa que vence en un tiempo específico en el futuro.

- **Capital:** Se ve como una opción de compra sobre los activos de la empresa con un precio de ejercicio igual al valor facial de la deuda.

**Lógica del Modelo:**

- Si, al vencimiento de la deuda, el valor de los activos de la empresa (V) es mayor que el valor de la deuda (D), la empresa puede pagar su deuda y los accionistas conservan la diferencia (V - D).

- Si el valor de los activos es menor que la deuda, la empresa incumple y los accionistas no reciben nada. En este caso, la empresa es entregada a los acreedores.

**Black & Scholes:**

**DP: Default probability.**

- La probabilidad de que el valor de los activos caiga por debajo del valor facial de la deuda en el momento del vencimiento. Esta PD se calcula utilizando la teoría de opciones, específicamente la fórmula Black-Scholes.
 
**Book Value Relationship:**

- A company balance sheet records book values—the value of a firm's equity E, its total assets A, and its total liabilities L. The relationship between these values is defined by the equation:

    - $A=E+L$

        - A = Assets

        - E = Equity

        - L = Liabilities

**Some real world problems and observations:**

- **Observable:** Equity is always observable, and is given by the firm's stock market price times the number of outstanding shares.

- **Unobservable:** Assets and Liabilities are reported infrequently, that's why they are called unobservable.

**Recap on how it's an option:**

**Option Pricing framework:**

- The Merton model assumes a single liability L with maturity T, usually a period of one year or less. At time T, the firm’s value to the shareholders equals the difference A – L when the asset value A is greater than the liabilities L. However, if the liabilities L exceed the asset value A, then the shareholders get nothing.

    - $E_T=max⁡(A_T-L,0)$
 
**Liability thershold:**

- In practice, firms have multiple maturities for their liabilities, so for a selected maturity T, a liability threshold L is chosen based on the whole liability structure of the firm. The liability threshold is also referred to as the default point. For a typical time horizon of one year, the liability threshold is commonly set to a value between the value of the short-term liabilities and the value of the total liabilities.

- Assuming a lognormal distribution for the asset returns, you can use the Black-Scholes-Merton equations to relate the observable market value of equity E, and the unobservable market value of assets A, at any time prior to the maturity T:

- WHY USE LOGNORMAL DISTRIBUTION?

**Lognormal distribution:**

- **Non-Negativity:** Ensures financial values (e.g., stock prices) are always positive.

- **Asymmetry:** Captures skewed financial returns with potential large outliers.

- **Multiplicative Nature:** Ideal for modeling percentage-based changes in stock prices. 

- **Heavy Tails:** Represents frequent extreme values better than the Normal distribution.

- **Historical Precedence:** Used in foundational financial models like Black-Scholes.

- **Limitations of Normal Distribution:** Inadequate for capturing unique financial data characteristics.

**Use the Black-Scholes-Merton:**

- Assuming a lognormal distribution for the asset returns, you can use the Black-Scholes-Merton equations to relate the observable market value of equity E, and the unobservable market value of assets A, at any time prior to the maturity T:
    
    - $E=AN(d_1 )-Le^(-rT) N(d_2 )$
    
- In this equation, r is the risk-free interest rate, N is the cumulative standard normal distribution, and d1 and d2 are given by:

    - $d_1=(ln⁡(A/L)+(r+0.5σ_A^2 )T)/(σ_a √T)$
    - $d_2=d_1-σ_a √T$

**Merton By Time Series:**

- If the equity time series has n data points, this approach calibrates a time series of n asset values A1 ,…,An that solve the following system of equations:

    - $E_1=A_1 N(d_1 )-L_1 e^(-r_1 T_1 ) N(d_2 )$

    - $E_2=A_2 N(d_1 )-L_2 e^(-r_2 T_2 ) N(d_2 )$

- The function directly computes the volatility of assets σA from the time series A1 ,…,An as the annualized standard deviation of the log returns. This value is a single volatility value that captures the volatility of the assets during the time period spanned by the time series.

**Distance to Default (DD):**

- After computing the values of A and σA, the function computes the distance to default (DD) is computed as the number of standard deviations between the expected asset value at maturity T and the liability threshold:

    - $DD=(ln⁡(A)+(μ_A-(σ_A^2)/2)T-ln⁡(L))/(σ_A √T)$

- The drift parameter μA is the expected return for the assets, which can be equal to the risk-free interest rate, or any other value based on expectations for that firm.

**Distance to Default:**

- A lower distance to default indicates a higher prob. default.

- $DD=(ln⁡(V/D)+(r+σ^2/2)T)/(σ√T)$

    - Where:
        - DD represents the distance to the default.
        
        - V is the market value of a company’s assets.

        - D is the market value of a company’s debt.

        - r is the risk-free interest rate.

        - σ is the volatility of the company’s asset value.

        - T is the time to maturity of the debt.

**Probability of default (PD):**

- Probability of the asset value failing below the liability thershold.

- $PD=1-N(DD)$

    - Where:
        
        - DD represents the distance to the default.

**Example #1:**

- Let’s consider an imaginary company called ABC Corporation, and we’ll use the Merton Model to estimate its credit risk and calculate the distance to default (DD) and probability of default (PD).

- ABC Corporation’s assets (V) market value: $100 million. ABC Corporation’s debt (D) market value is $60 million. Risk-f ree interest rate (r): 3%. The volatility of ABC Corporation’s asset value (σ): 0.2. Time to maturity of the debt (T): 2 years.

- In this example, the calculated distance to default (DD) for ABC Corporation is approximately 2.15. The DD represents the company’s buffer or cushions before it reaches the default threshold. A higher DD indicates a lower probability of default, suggesting that ABC Corporation has a relatively lower credit risk. The PD is 1.5% more or less.

**Assumptions:**

- **Continuous-time framework:** which considers the value of a company’s assets and debt continuously evolving.

- **Geometric Brownian motion:** company’s assets follows a geometric Brownian movement.

- **Tradeable debt and equity:** the debt and equity of a company are tradeable.

- **Risk-neutral valuation:** it assumes market participants are riskneutral and make decisions based on riskfree interest rates.

- **No taxes or transaction costs:** absence of taxes and transaction costs.

- **Constant asset volatility:** the volatility of a company’s asset value (σ) remains consistent over time.
